```python
from torch.nn.utils.rnn import pad_sequence
class dataset_pos_encoder(Dataset):

    def __init__(self,words,tags,device):
        
        self.words=words
        self.tags=tags
        self.device=device

    def __len__(self):
        return len(self.words)
    

    def feature_matrix(self,words):
        f_mat=[]
        for i in words:
            try:
                f_mat.append(torch.Tensor(find_closest_embedding[i]).unsqueeze(dim=0))

            except:

                x=np.random.normal(loc=0.0,scale=0.1,size=300)
                f_mat.append(torch.Tensor(x).unsqueeze(dim=0))
        f_mat=torch.cat(f_mat,axis=0).to(self.device)
        return f_mat

    def __getitem__(self, idx):
        # sentence=[]
        # tag_list=[]
        words=self.words[idx]
        tags=self.tags[idx]
        # tagged_sent=self.data[idx]
        # for item in tagged_sent:
            # sentence.append(item[0])
            # tag_list.append(item[1])
        feature=self.feature_matrix(words)
        return (feature), torch.Tensor([int(tag_dictionary[i]) for i in tags])
```

```python
data_train=dataset_pos_encoder(train_words,train_tags,"cuda")
def pad_collate(batch):
    (xx, yy) = zip(*batch)
    x_lens = [len(x) for x in xx]
    y_lens = [len(y) for y in yy]

    xx_pad = pad_sequence(xx, batch_first=True, padding_value=11)
    yy_pad = pad_sequence(yy, batch_first=True, padding_value=11)

    return xx_pad, yy_pad
dataloader = DataLoader(data_train, batch_size=100,shuffle=True, num_workers=0, collate_fn=pad_collate)
```

```python
import torch.nn as nn
class encoder(nn.Module):
    def __init__(self,hidden_dim=50,layer_dim=1,in_dim=100,out_dim=12,device="cuda"):
        super().__init__()
        self.hidden_dim=hidden_dim
        self.layer_dim=layer_dim
        self.in_dim=in_dim
        self.out_dim=out_dim
        self.device=device
        self.rnn1 = nn.RNN(in_dim, hidden_dim, layer_dim, batch_first=True, nonlinearity='relu',bidirectional=True).to(device)        
        self.softmax=nn.Softmax(dim=-1).to(device)
        self.lin=nn.Linear(2*hidden_dim,out_dim).to(device)

    def forward(self,x):
        x=x.to(self.device)
        out,hn=self.rnn1(x)
        x=self.lin(out)
        x=self.softmax(x)
        return x
```

```python
def train_func(data_train, data_test):
    import time
    device="cuda"
    dataloader = DataLoader(data_train, batch_size=100,
                            shuffle=True, num_workers=0,collate_fn=pad_collate)
    model=encoder(device=device)
    optimizer = torch.optim.SGD(model.parameters(), lr=1e-3, momentum=0.9)
    loss_fn=nn.CrossEntropyLoss()
    model.train()
    start=time.time()
    num_epochs=5
    for e in range(num_epochs):

        for batch, (X, y) in enumerate(dataloader):
            size = len(dataloader.dataset)
            y_flat = y.flatten()
            y_hot = torch.zeros(y_flat.size(), 3)
            y_hot[np.arange(y_flat.size()), y_flat] = 1
            y = y_hot.reshape(y.size())
            X, y = X.to(device), y.to(device)

            # Compute prediction error
            pred = model(X)
            print(pred.shape)
            print(y.shape)

            loss = loss_fn(pred, y)

            # Backpropagation
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            if batch % 500 == 0:
                end=time.time()
                loss, current = loss.item(), (batch + 1) * len(X)
                print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
                print("time: ", end-start)
            break
        break
    
    

    dataloader = DataLoader(data_test, batch_size=100,
                            shuffle=True, num_workers=0)
       
    a = torch.tensor([]).to(device)
    b = torch.tensor([]).to(device)

    for batch, (X, y) in enumerate(dataloader):
        size = len(dataloader.dataset)

        X, y = X.to(device), y.to(device)


        # Compute prediction error
        pred = model(X)
        print('pred shape',pred.shape)
        print('y shape',y.shape)
        loss = loss_fn(pred, y)
        a = torch.cat((a, y))
        b = torch.cat((b, torch.argmax(pred, axis = 1)))

    return a, b
```
